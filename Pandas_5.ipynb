```python
dire=pd.read_csv("directors.csv",index_col=0)
mov=pd.read_csv("movies.csv",index_col=0)
```

```python
dire.head()
```

```python
mov.head()
```

```python
data=mov.merge(dire,how="left",left_on="director_id",right_on="id")
```

```python
data
```

```python
data["budget"]=(data["budget"]/1000000).round(2)
```

```python
def calc_exp(x):
    x=x-x.mean()
    return x


data.groupby("director_name")["budget"].transform(calc_exp)
```

```python
data
```

```python
def cal_risk(x):
    x["risky"]=(x["budget"]-x["revenue"].mean())<=0
    return x

data_risky=data.groupby("director_name").apply(cal_risk)
data_risky
```
