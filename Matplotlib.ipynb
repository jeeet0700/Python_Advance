```python
x=np.arange(-3,3,0.1)
y=x**2
y1=x**3
y2=x**4
plt.plot(x,y,label="Sq")
plt.plot(x,y1,label="Cube")
plt.plot(x,y2,label="pow 4")
plt.legend()
plt.show()
```

```python
x=np.arange(-3,3,0.1)
y=x**2
y1=x**3
y2=x**4
plt.plot(x,y)
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(["Sq","Cube","pow 4"])
plt.show()
```

```python
x=np.arange(-3,3,0.1)
y=x**2
y1=x**3
y2=x**4
plt.plot(x,y,label="Sq")
plt.plot(x,y1,label="Cube")
plt.plot(x,y2,label="pow 4")
plt.legend(loc="upper center")
plt.show()
```

```python
x=np.arange(-3,3,0.1)
y=x**2
y1=x**3
y2=x**4
plt.plot(x,y,label="Sq")
plt.plot(x,y1,label="Cube")
plt.plot(x,y2,label="pow 4")
plt.grid()
plt.legend(loc="upper center",frameon=False)
plt.show()
```
