```python
x=np.arange(-3,3,0.1)
y=x**3
plt.xticks(np.arange(-3,3,0.2),rotation=90)
plt.yticks(np.arange(-50,50,3))
plt.title("X**3")
plt.xlabel("x-value")
plt.ylabel("y-value")
plt.plot(x,y)
plt.show()
```
