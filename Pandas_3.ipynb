```python
user=pd.DataFrame({"userId":[1,2,3],"name":["Ram","Sham","Aam"],})
```

```python
user
```

```python
msg=pd.DataFrame({"userId":[1,1,2,4],"msg":["hi","hello","hmm","kool"]})
```

```python
msg
```

```python
user.merge(msg,on="userId")
```

```python
pd.merge?
```

```python
user.merge(msg,on="userId",how="outer")
```

```python
user.merge(msg,on="userId",how="right")
```

```python
user.merge(msg,on="userId",how="left")
```

```python
user.rename(columns={"userId":"user_ud"},inplace=True)
```

```python
user
```

```python
user.merge(msg,left_on="user_ud",right_on="userId")
```
