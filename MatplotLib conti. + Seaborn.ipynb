```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
val=[1,1,2,4,6,4,2,8,1,3,5]
n,bins,_ = plt.hist(val)
plt.show()
```

```python
print(n)
print(bins)
```

```python
val=[1,1,2,4,6,4,2,8,1,3,5]
n,bins,_ = plt.hist(val,bins=[1,2,3,4,5])
plt.show()
```

```python
print(n)
print(bins)
```

```python
x=np.arange(0,20)
y=np.linspace(1,20,20,dtype=np.int32)+np.random.rand(20)*10
print(x)
print(y)
plt.plot(x,y)
```

```python
x=np.arange(0,20)
y=np.linspace(1,20,20,dtype=np.int32)+np.random.rand(20)*10
print(x)
print(y)
plt.scatter(x,y)
```

```python
x=np.arange(0,20)
y=np.linspace(1,20,20,dtype=np.int32)+np.random.rand(20)*10
y1=np.linspace(1,20,20,dtype=np.int32)+np.random.rand(20)*20
plt.scatter(x,y,color="g",s=4*abs(y))
plt.scatter(x,y1,color="c",s=4*abs(y1))
plt.show()
```
